```python
df.Date = pd.to_datetime(df.Date)
df
```

```python
##info about holiday
SuperBowl = ['12-2-2010', '11-2-2011', '10-2-2012']
LabourDay =  ['10-9-2010', '9-9-2011', '7-9-2012']
ThanksGiving =  ['26-11-2010','25-11-2011', '23-11-2012']
Christmas = ['31-12-2010', '30-12-2011', '28-12-2012']
# Calculating holiday events sales
SuperBowlSales = round(df.loc[df.Date.isin(SuperBowl)]['Weekly_Sales'].mean(),2)
LabourDaysales = round(df.loc[df.Date.isin(LabourDay)]['Weekly_Sales'].mean(),2)
ThanksgivingSales = round(df.loc[df.Date.isin(ThanksGiving)]['Weekly_Sales'].mean(),2)
ChristmasSales = round(df.loc[df.Date.isin(Christmas)]['Weekly_Sales'].mean(),2)
SuperBowlSales,LabourDaysales,ThanksgivingSales,ChristmasSales
```

```python
nonHolidaySales = round(df.query('Holiday_Flag == 0')['Weekly_Sales'].mean(),2)
nonHolidaySales
```

```python
result = {'Super Bowl Sales':SuperBowlSales,
              'Labour Day Sales':LabourDaysales,
              'Thanksgiving Sales':ThanksgivingSales,
              'Christmas Sales':ChristmasSales,
              'Non Holiday Sales':nonHolidaySales}

result
```

```python
for i in result:
    if(result[i]>nonHolidaySales):
        print('{} higher than the mean sales in the non-holiday season'.format(i))
```

```python
c=['red','red','red','red','orange']
plt.figure(figsize=(10,8))
plt.bar(result.keys(), result.values(), color =c,width = 0.4,alpha=0.9);
```

```python
# 2010 
dx=df.set_index('Date').sort_values('Date')
spring=dx.loc['2010-03-20':'2010-06-20'].Weekly_Sales.sum()
summer=dx.loc['2010-06-21':'2010-09-22'].Weekly_Sales.sum()
autumn=dx.loc['2010-09-23':'2010-12-21'].Weekly_Sales.sum()
winter=dx.loc['2010-12-22':'2011-03-20'].Weekly_Sales.sum()

dxx=pd.DataFrame({"spring":spring,
                  "summer":summer,
                  "autumn":autumn,
                  "winter":winter},index=[0])
sns.set_style("whitegrid")
sns.barplot(data=dxx).set(xlabel ="Semester", ylabel = "Weekly Sales", title ='2010');
```
